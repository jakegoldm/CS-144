```python
import networkx as nx
```

```python
G = nx.karate_club_graph()
nx.draw(G, pos=nx.kamada_kawai_layout(G), with_labels=True)
```

### Degree Centrality

```python
degs = sorted(nx.degree_centrality(G).items(), key=lambda x:x[1], reverse=True)
degs[:5]
```

### Closeness Centrality

```python
cls = sorted(nx.closeness_centrality(G).items(), key=lambda x:x[1], reverse=True)
cls[:5]
```

### Betweenness Centrality

```python
bet = sorted(nx.betweenness_centrality(G).items(), key=lambda x:x[1], reverse=True)
bet[:5]
```

### PageRank

```python
prs = sorted(nx.pagerank(G).items(), key=lambda x:x[1], reverse=True)
prs[:5]
```

Degree, Closeness, and Betweenness Centrality yield different rankings as shown by the examples above. Therefore, it is not necessary to show case-by-case that each pair is different because they all produce different orderings for the top 5 nodes in the network. Note that the PageRank and Degree rankings are proportional. We prove that the actual computed centralities for both Degree and PageRank are proportional below. 

To prove that $C_D (i) \propto r_i$ for a node $i$, we show that there exists some constant $k$ such that $r_i = k d_i$. Observe that if $$k = \frac{1}{\sum_{i = 1}^n d_i},$$ which is constant, then $$
r_i = \sum_{j \in N(i)} \frac{r_j}{d_j} = \sum_{j \in N(i)} \frac{k d_j}{d_j} = \sum_{j \in N(i)} k = k d_i.$$

QED. 

## Part b

Degree Centrality:

Social Network Analysis
Degree centrality measures the number of connections a node has in a network. In a social network setting, nodes with high degree centrality represent individuals who are highly connected to others.

Closeness Centrality:

Transportation Network Analysis
Closeness centrality measures how close a node is to all other nodes in the network. In a transportation network analysis, closeness centrality can identify locations (nodes) that are central in terms of accessibility and travel time to other locations. 

Betweenness Centrality:

Infrastructure Network Analysis
Betweenness centrality quantifies the number of times a node acts as a bridge along the shortest path between two other nodes. In an infrastructure network analysis, such as analyzing power grids or internet networks, betweenness centrality identifies nodes that maintain power. 

PageRank Centrality:

Web Page Ranking
PageRank centrality measures the importance of a node based on the number and quality of links pointing to it. In the context of web page ranking, PageRank is used by search engines to rank web pages in their search results. Pages with higher PageRank centrality are considered more important because they are linked to by other important pages.
